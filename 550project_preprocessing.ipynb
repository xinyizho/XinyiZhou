```python
for i in dates:
  high = pd.to_datetime(i) + relativedelta(months =1)
  low = pd.to_datetime(i) - relativedelta(months =1)
  a = s[s["Date"] >= low.strftime('%Y-%m-%d')]
  a = a[a["Date"] <= high.strftime('%Y-%m-%d')]
  spc = SpCopy[SpCopy["Date"] == i]
  spc = spc[["Title", "Artists"]].drop_duplicates()
  a = a[["Song", "Artist", "id"]].drop_duplicates()
  res = spc.merge(a, left_on = "Title", right_on = "Song")
  for i in res.index:
    if(jellyfish.jaro_similarity(res.loc[i,"Artists"], res.loc[i, "Artist"]) > 0.6):
      sp_reduced.loc[(sp_reduced["Title"] == res.loc[i, "Title"]) &
               (sp_reduced["Artists"] == res.loc[i, "Artists"]), "id"] = res.loc[i, "id"]
```
